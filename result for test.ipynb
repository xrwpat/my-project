```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
result=pd.read_excel('/Users/xie/Downloads/Qishon/订单路由问题/result_10000.xlsx')
store=pd.read_excel('/Users/xie/Downloads/Qishon/订单路由问题/数据/仓库档案.xlsx')
order=pd.read_excel('/Users/xie/Downloads/Qishon/订单路由问题/数据/零售小票.xlsx')
stock=pd.read_excel('/Users/xie/Downloads/Qishon/订单路由问题/数据/库存数据.xlsx')
```

```python
def dropsomestr(x):
    x=x.replace('\n','')
    x=x.replace('.',',')
    x=x.replace('[','')
    x=x.replace(']','')
    x=x.replace(',','')
    x=x.replace(' ','')
    x=np.array([int(i) for i in list(x)])
    x=np.reshape(x,(-1,30))
    return(x)
result['发货情况']=result['发货情况'].apply(dropsomestr)
result['size']=[x.size for x in result['发货情况']]
```

```python
result.reset_index(inplace=True)
result
```

```python
order_for_test=locals()
s=0
for i in result[(result['size']>60)]['index']:
    s=s+1
    print(s,i)
    order_for_test[i]=pd.DataFrame(store)
    index_num=result[(result['index']==i)].index.tolist()
    order_for_test[i+'_result']=pd.DataFrame(result['发货情况'][index_num[0]]).T
    order_for_test[i]=pd.concat([order_for_test[i+'_result'],order_for_test[i]],axis=1)
    address=order[(order['单据编号']==i)]['收货地址'].reset_index()
    order_for_test[i]['收货地址']=address['收货地址'][0]
    order_num=order[(order['单据编号']==i)]['数量']
    for j in order[(order['单据编号']==i)]['商品代码']:
        count=order[(order['单据编号']==i)&(order['商品代码']==j)].index.tolist()
        count=count[0]
        order_num_new=order_num[count]
        order_for_test[i]['下单数'+'{}'.format(j)]=order_num_new
        order_for_test[i]['库存数'+'{}'.format(j)]=0
        for k in order_for_test[i].index:
            a=order_for_test[i]['仓库名称'][k]
            b=stock[(stock['商品代码']==j) & (stock['仓库名称']==a)]
            order_for_test[i]['库存数'+'{}'.format(j)][k]=b['数量'].sum()
    order_for_test[i].to_csv('/Users/xie/Downloads/Qishon/订单路由问题/数据/result'+'{}'.format(i)+'.csv')
```
